```python
import pandas as pd
import numpy as np
df = pd.read_csv('StockUS.csv')
df
```

```python
def growth_rate_revenue_Q3_to_Q2(df):
    Q2_minus_Q3 = df['Total Revenue Q2'] - df['Total Revenue Q3']
    return abs(Q2_minus_Q3 / df['Total Revenue Q3'] if df['Total Revenue Q3'] != 0 else Q2_minus_Q3)
    
def growth_rate_revenue_Q2_to_PQ(df):
    PQ_minus_Q2 = df['Total Revenue PQ'] - df['Total Revenue Q2']
    return abs(PQ_minus_Q2 / df['Total Revenue Q2'] if df['Total Revenue Q2'] != 0 else PQ_minus_Q2)

# ///////////////////////

def growth_rate_net_income_Q3_to_Q2(df):
    Q2_minus_Q3 = df['Net Income Q2'] - df['Net Income Q3']
    return abs(Q2_minus_Q3 / df['Net Income Q3'] if df['Net Income Q3'] != 0 else Q2_minus_Q3)

def growth_rate_net_income_Q2_to_PQ(df):
    PQ_minus_Q2 = df['Net Income PQ'] - df['Net Income Q2']
    return abs(PQ_minus_Q2 / df['Net Income Q2'] if df['Net Income Q2'] != 0 else PQ_minus_Q2)

# ///////////////////////
#calculate the momentum in the change rate of financial indices reports

#calculate the absolute change rate revenue from the previous 3rd quarter to the previous 2nd quarter and
#the absolute change revenue from the previous 2nd quarter to the previous quarter and
#compare them
df['growth_rate_revenue_Q3_to_Q2'] = df.apply(growth_rate_revenue_Q3_to_Q2, axis=1)
df['growth_rate_revenue_Q2_to_PQ'] = df.apply(growth_rate_revenue_Q2_to_PQ, axis=1)
df['growth_rate_revenue_Q2PQ_compare_Q3Q2'] = abs(df['growth_rate_revenue_Q2_to_PQ'] - df['growth_rate_revenue_Q3_to_Q2'])

#calculate the absolute change net income from the previous 3rd quarter to the previous 2nd quarter and
#the absolute change net income from the previous 2nd quarter to the previous quarter and
#compare them
df['growth_rate_net_income_Q3_to_Q2'] = df.apply(growth_rate_net_income_Q3_to_Q2, axis=1)
df['growth_rate_net_income_Q2_to_PQ'] = df.apply(growth_rate_net_income_Q2_to_PQ, axis=1)
df['growth_rate_net_income_Q2PQ_compare_Q3Q2'] = abs(df['growth_rate_net_income_Q2_to_PQ'] - df['growth_rate_net_income_Q3_to_Q2'])

#calculate the absolute change EPS from the previous 3rd quarter to the previous 2nd quarter and
#the absolute change EPS from the previous 2nd quarter to the previous quarter and
#compare them
df['growth_change_EPS_Q3_to_Q2'] = abs(df['Basic EPS Q2'] - df['Basic EPS Q3'])
df['growth_change_EPS_Q2_to_PQ'] = abs(df['Basic EPS PQ'] - df['Basic EPS Q2'])
df['growth_change_EPS_Q2PQ_compare_Q3Q2'] = abs(df['growth_change_EPS_Q2_to_PQ'] - df['growth_change_EPS_Q3_to_Q2'])

#calculate the absolute change ROE from the previous 3rd quarter to the previous 2nd quarter and
#the absolute change ROE from the previous 2nd quarter to the previous quarter and
#compare them
df['growth_change_ROE_Q3_to_Q2'] = abs(df['ROE Q2'] - df['ROE Q3'])
df['growth_change_ROE_Q2_to_PQ'] = abs(df['ROE PQ'] - df['ROE Q2'])
df['growth_change_ROE_Q2PQ_compare_Q3Q2'] = abs(df['growth_change_ROE_Q2_to_PQ'] - df['growth_change_ROE_Q3_to_Q2'])

#calculate the absolute change RSI from the previous 3rd quarter to the previous 2nd quarter and
#the absolute change RSI from the previous 2nd quarter to the previous quarter and
#compare them
df['RSI_Q3Q2'] = abs(df['RSI_Q2'] - df['RSI_Q3'])
df['RSI_Q2PQ'] = abs(df['RSI_PQ'] - df['RSI_Q2'])
df['RSI_Q2PQ_Q3Q2'] = abs(df['RSI_Q3Q2'] - df['RSI_Q2PQ'])
```

```python
#filter columns for EDA
df_out = df.drop(columns=['Name', 'Total Revenue Q3', 'Total Revenue Q2', 'Total Revenue PQ',
                         'Net Income Q3', 'Net Income Q2', 'Net Income PQ',
                         'ROE Q3', 'ROE Q2', 'ROE PQ'], axis=1)
df_out = df_out[[c for c in df_out if c not in ['Highest Price in Next Quarter', 'Lowest Price in Next Quarter']]
                + ['Highest Price in Next Quarter', 'Lowest Price in Next Quarter']]
df_out
```

```python
df_out.to_csv('stock_us_cleaned.csv', index=False)
```
